```python
salary = float(input("Enter your salary: "))
years = int(input("Enter your years of service: "))

if years > 5:
    bonus = salary * 0.05
else:
    bonus = 0

print("Net bonus amount:", bonus)
```

```python
# Program to check whether a number is even or odd

num = int(input("Enter a number: "))

if num % 2 == 0:
    print("The number is Even.")
else:
    print("The number is Odd.")
```

```python
# Program to check whether a number is divisible by 7 or not

num = int(input("Enter a number: "))

if num % 7 == 0:
    print("The number is divisible by 7.")
else:
    print("The number is not divisible by 7.")
```

```python
# Program to display "Hello" if the number is a multiple of 5,
# otherwise display "Bye"

num = int(input("Enter a number: "))

if num % 5 == 0:
    print("Hello")
else:
    print("Bye")
```

```python
length = float(input("Enter the length: "))
breadth = float(input("Enter the breadth: "))

if length == breadth:
    print("It is a Square.")
else:
    print("It is a Rectangle.")
```

```python
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Check and print the greatest number
if num1 > num2:
    print("Greatest number is:", num1)
elif num2 > num1:
    print("Greatest number is:", num2)
else:
    print("Both numbers are equal.")
```
